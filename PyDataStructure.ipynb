20. Explain the difference between a list and a dictionary in terms of data retrieval.
---
- Lists retrieve data using index-based access (O(1)), while dictionaries use key-based lookups (O(1) on average, faster for large data sets), making lists better for ordered sequences and dictionaries ideal for fast key-value retrieval.
